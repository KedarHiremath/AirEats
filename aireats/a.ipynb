## 🔹 Frontend: React.js + TailwindCSS / Bootstrap

| Pros                                                                                   | Cons                                                                                                                                    |
| -------------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------- |
| • Very mature, lots of community & libraries; fast dev with reusable components.       | • React has a learning curve if someone isn’t familiar (state management, hooks, etc.).                                                 |
| • Tailwind / Bootstrap speed up styling a lot; Tailwind especially gives more control. | • Tailwind’s utility-class approach might get messy if styles not well organized. Bootstrap can make UI look generic unless customized. |
| • Good for embedding chatbot widget, responsive web portals.                           | • Performance concerns if too many re-renders; on low-end devices, responsiveness may lag unless optimized.                             |
| • Lots of tool-ing: dev servers, hot reload, testing, etc.—benefits SIH speed.         | • If you want mobile app later, React alone may need to integrate separately; cross platform sometimes adds complexity.                 |

---

## 🔹 Backend: Django + DRF, optional FastAPI microservice

| Pros                                                                                                          | Cons                                                                                                                                        |
| ------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------- |
| • Django gives you a strong, tested ecosystem: user management, ORM, admin panel, etc. Speeds up boilerplate. | • Django can be heavier; for extremely lightweight endpoints or real-time features might have overhead.                                     |
| • DRF makes building REST APIs structured; serializers, permissions, etc.                                     | • Serialization & DB queries can get slow if data / relationships not optimized.                                                            |
| • FastAPI as a microservice for NLP tasks gives you speed & async capabilities.                               | • Having more than one backend stack (Django + FastAPI) means more infrastructure, deployment, coordination work. More to setup & maintain. |
| • Good Python ecosystem around NLP so alignment with tools is easier.                                         | • Adding microservices increases complexity for data sharing, versioning, deployment.                                                       |

---

## 🔹 Database: PostgreSQL (or SQLite for demo)

| Pros                                                                                                   | Cons                                                                                                                 |
| ------------------------------------------------------------------------------------------------------ | -------------------------------------------------------------------------------------------------------------------- |
| • PostgreSQL is reliable, supports complex queries, scales well. Free tiers available (e.g. Supabase). | • For hackathon setup, config might take time; also indexing / schema design matters. Mistakes lead to slow queries. |
| • Good support for JSON fields if needed (for flexible template data).                                 | • If data is mostly unstructured (text, logs), using only SQL might get cumbersome; may need hybrid approach.        |
| • SQLite is super simple for prototyping / demo.                                                       | • SQLite doesn’t scale; concurrency / large data loads, multi-user heavy load will stress it.                        |

## 🔹 NLP / AI (Rasa, IndicTrans2, Sentence-BERT + FAISS, PDF extractors)

| Pros                                                                                                                     | Cons                                                                                                                                      |
| ------------------------------------------------------------------------------------------------------------------------ | ----------------------------------------------------------------------------------------------------------------------------------------- |
| • Rasa is open-source, fairly robust for intent recognition + context + fallback. You can also build template responses. | • Need nontrivial setup (defining intents, training data, pipelines). If training data is small, performance may suffer.                  |
| • IndicTrans2 and open HF models provide free multilingual translation; better coverage for regional languages.          | • Latency might be higher for large models; resource constraints (RAM, GPU) may limit performance.                                        |
| • Sentence-BERT + FAISS gives you semantic search; better than keyword matching for circulars / PDFs.                    | • Indexing (of many documents) takes effort; keeping it updated when new circulars are added is work. Also model sizes may cause latency. |
| • PDF extraction tools (PyPDF2, pdfplumber) allow ingestion of circulars etc. into text form.                            | • PDF parsing is messy: layout, tables, images → often ambiguous; cleaning / error handling needed.                                       |
| • All open-source / free so fits hackathon + budget.                                                                     | • Maintenance: models need updates; there may be edge cases (language variants, typos, colloquial speech) that these tools struggle with. |

## 🔹 Voice + Multilingual Support (Vosk / Whisper / gTTS etc.)

| Pros                                                                                                             | Cons                                                                                                                                                                  |
| ---------------------------------------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| • Vosk (offline) helps maintain privacy, works without net or with poor connection. Whisper gives high accuracy. | • Whisper and other speech-to-text models are large; will need good computational resources (GPU if possible) or accept slower response times.                        |
| • gTTS / pyttsx3 for TTS are simple to integrate; good for basic responses.                                      | • TTS voices may sound robotic; regional accents / pronunciations may be off. Also latency and network issues may affect user experience.                             |
| • Using free translation + multilingual models helps reach regional language users.                              | • Translation models sometimes misinterpret administrative / domain-specific jargon; risk of wrong meaning. Also handling mixing languages (code-mixing) gets tricky. |

## 🔹 Admin Side: Django Admin + React Dashboard + Template System

| Pros                                                                                                                  | Cons                                                                                                                                                                   |
| --------------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| • Django Admin gives you a ready-to-use interface for permissions, CRUD, etc. Staff can use it with minimal training. | • Django Admin’s UI is basic; for nontechnical staff you might need more polished dashboard (React) – double work.                                                     |
| • React Dashboard allows more custom UX, visual clarity (templates, analytics, logs).                                 | • Maintaining two UIs (React + Django Admin) increases work. Need to sync permissions, data validation across both.                                                    |
| • Template system means clerical staff can enter info without needing developers each time. Speeds updates.           | • Template design needs to be flexible but well-validated; if you allow too much freedom, data can become inconsistent (fields missing, wrong types etc.), hurting UX. |
